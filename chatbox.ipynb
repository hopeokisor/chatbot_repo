```python
import nltk
import random
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity      
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
# import spacy
lemmatizer = nltk.stem.WordNetLemmatizer()
# Download required NLTK data
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
```

```python
df = pd.read_csv('Mental_Health_FAQ.csv', na_filter=False)
df = df[['Questions', 'Answers']]
df.head(10)
```

```python
sent_tokens = []
for i in df['Questions']:
    sent_tokens.append(nltk.sent_tokenize(i))

sent_tokens    
```

```python
word_tokens = []
for i in sent_tokens:
    word_tokens.append(nltk.word_tokenize(str([i for i in sent_tokens])))

word_tokens[1]  
```

```python
temporal = [lemmatizer.lemmatize(i.lower()) for i in word_tokens[1]]
temporal2 = ' '.join(temporal)
preprocessed = []
preprocessed.append(temporal2)
preprocessed
```

```python

# Define a function for text preprocessing (including lemmatization)
def preprocess_text(text):
    
    # Identifies all sentences in the data
    sentences = nltk.sent_tokenize(text)
    
    # Tokenize and lemmatize each word in each sentence
    preprocessed_sentences = []
    for sentence in sentences:
        tokens = [lemmatizer.lemmatize(word.lower()) for word in nltk.word_tokenize(sentence) if word.isalnum()]
        # Turns to basic root - each word in the tokenized word found in the tokenized sentence - if they are all alphanumeric 
        # The code above does the following:
        # Identifies every word in the sentence 
        # Turns it to a lower case 
        # Lemmatizes it if the word is alphanumeric

        preprocessed_sentence = ' '.join(tokens)
        preprocessed_sentences.append(preprocessed_sentence)
    
    return ' '.join(preprocessed_sentences)


df['tokenized Questions'] = df['Questions'].apply(preprocess_text)
df.head()
```

```python
df['preprocessed_text'] = df['Questions'].apply(preprocess_text)
df.head()
```

```python
input_text = input('input your text: ')
input_text
```

```python
corpus = df['preprocessed_text'].to_list()
corpus
```

```python
# Vectorize corpus
tfidf_vectorizer = TfidfVectorizer()
vectorized_corpus = tfidf_vectorizer.fit_transform(corpus)
print(vectorized_corpus)
```

```python
preprocessed_input_text = preprocess_text(input_text)
```

```python
transformed_preprocessed_input_text = tfidf_vectorizer.transform([preprocessed_input_text])
print(transformed_preprocessed_input_text)
```

```python
similarity_score = cosine_similarity(transformed_preprocessed_input_text, vectorized_corpus)
print(similarity_score)
```

```python
most_similar_word = similarity_score.argmax()
```

```python
chatbot_reply = df['Answers'].iloc[most_similar_word]
print(chatbot_reply)
```
